# Welcome to Colab!

```
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('/content/iris.csv')

# Tampilkan 5 data pertama
data.head()
```

```
# Informasi dataset
data.info()
```

```
# Jumlah data dan fitur
print("Jumlah data:", data.shape[0])
print("Jumlah fitur:", data.shape[1] - 1)  # dikurangi target
```

```
# Distribusi kelas target
data['virginica'].value_counts()
```

```
# Visualisasi distribusi kelas
sns.countplot(x='virginica', data=data)
plt.title("Distribusi Kelas Target")
plt.show()
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```
# Pisahkan fitur dan target
X = data.drop('virginica', axis=1)
y = data['virginica']

# Split data 70% training dan 30% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```
# Scaling fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```
# Tentukan nilai K
k = 5
knn = KNeighborsClassifier(n_neighbors=k)

# Latih model
knn.fit(X_train_scaled, y_train)

# Prediksi
y_pred = knn.predict(X_test_scaled)
```

```
# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```
# Precision, Recall, F1-score
print(classification_report(y_test, y_pred))
```
